```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Diwali Sales Data.csv', encoding='unicode_escape')
```

```python
df.head(10)
```

```python
df=df.drop(['Status','unnamed1'], axis=1)
```

```python
df.info()
```

```python
pd.isnull(df).sum()
```

```python
df.shape
```

```python
#drop null values
df.dropna(inplace=True)
```

```python
df.shape
```

```python
#change data type
df['Amount'] = df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

```python
df.columns
```

```python
#renmae column
df.rename(columns={'Marital_Status':'Shaadi'})
```

```python
df.describe()
```

```python
df[['Age','Orders','Amount']].describe()
```

```python
ax=sns.countplot(x='Gender', data=df)
ax.bar_label(ax.containers[0])
plt.show()
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sg=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=sg)
plt.show()
```

```python
sns.countplot(data=df,x='Age Group')
plt.show()
```

```python
ax=sns.countplot(data=df,x='Age Group',hue='Gender')
ax.bar_label(ax.containers[0])
plt.show()
```

```python
sa=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=sa)
plt.show()
```

```python
ss=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(25,10)})
sns.barplot(x='State',y='Orders',data=ss)
plt.show()
```

```python
sss=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(25,10)})
sns.barplot(x='State',y='Amount',data=sss)
plt.show()
```

```python
ax=sns.countplot(data=df,x='Marital_Status')
ax.bar_label(ax.containers[0])
sns.set(rc={'figure.figsize':(5,5)})
plt.show()
```

```python
sm=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(x='Marital_Status',y='Amount',data=sm,hue='Gender')
plt.show()
```

```python
ax=sns.countplot(data=df,x='Occupation')
ax.bar_label(ax.containers[0])
sns.set(rc={'figure.figsize':(25,25)})
plt.show()
```

```python
so=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(25,25)})
sns.barplot(x='Occupation',y='Amount',data=so)
plt.show()
```

```python
ax=sns.countplot(data=df,x='Product_Category')
ax.bar_label(ax.containers[0])
sns.set(rc={'figure.figsize':(5,5)})
plt.show()
```

```python
sp=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(35,35)})
sns.barplot(x='Product_Category',y='Amount',data=sp)
plt.show()
```

```python
si=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='Product_ID',y='Orders',data=si)
plt.show()
```
